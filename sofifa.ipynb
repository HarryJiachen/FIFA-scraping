```python
import requests
from bs4 import BeautifulSoup
from pandas.core.frame import DataFrame
import pandas as pd
```

```python
r = requests.get('https://sofifa.com/?col=pt&sort=desc', timeout=30)
r2 = requests.get('https://sofifa.com/?col=pt&sort=desc&offset=60', timeout=30)
```

```python
soup = BeautifulSoup(r.text, "html.parser")
soup2 = BeautifulSoup(r2.text, "html.parser")
```

```python
data1=soup.findAll('td')
data2=soup2.findAll('td')
```

```python
l=list()
```

```python
for i in data1:
      a=i.text
      l.append(a)
for i in data2:
      a=i.text
      l.append(a)
len(l)
```

```python
l2=list()
for j in range(120):
    l2.append(l[10*j+1:10*j+10])
```

```python
for i in range(120):
    ex=l2[i][0]
    ex2=l2[i][4]
    l2[i][0]=(ex.split('\n')[2])
    l2[i].insert(1,ex.split('\n')[3])
    l2[i][5]=(ex2.split('\n')[4])
    l2[i].insert(6,ex2.split('\n')[5])
    l2[i][-1]=l2[i][-1].split('\n',1)[1]
```

```python
column_names=['Name','Position','Age','Overall','Potential','Club','Contract','Value','Wage','Stats','Hits']
```

```python
l2.insert(0,column_names)
```

```python
data=DataFrame(l2)
data
```

```python
data.to_csv('sofifa.csv',encoding='utf_8_sig',index=False) 
```
